```python
# coding=utf8
import openpyxl
import re

# [\000-\010]|[\013-\014]|[\016-\037]
class xie_xlsx():
    def __init__(self,name,biandan='Sheet1'):
        self.name01=name
        self.biaodan=biandan

    # 读表
    def dubiao(self):
        workbook = openpyxl.load_workbook(self.name01)
        # sheet = wb.get_sheet_by_name(sheet_name)这种方式已经弃用，不建议使用
        sheet = workbook[self.biaodan]
        zhuan02=[]
        for row in sheet.rows:
            zhuan01=[]
            for cell in row:
                zhuan01.append(cell.value)
                # print(cell.value, "\t", end="")
            zhuan02.append(zhuan01)

        workbook.close()
        return zhuan02

    # 创建表
    def chuangbiao(self,tou):
        workbook = openpyxl.Workbook()
        sheet = workbook.active
        sheet.title =self.biaodan
        # 写入数据，从（1,1）开始
        # 写入数据，从（1,1）开始
        # sheet.cell(1, 1, value=str('5'))
        sheet.append(tou)
        workbook.save(self.name01)
        print("表格创建成功。。。。。。。")

    # 对表进行修改，追加
    def xiubiao02(self,data_list):
        wb = openpyxl.load_workbook(self.name01)
        # 切换到目标数据表
        ws = wb[self.biaodan]
        # 写入数据，从（1,1）开始
        # ws.cell(1,1,'ass')
        # ws.cell(row=1, column=1).fill = sty.PatternFill(fill_type='solid', fgColor="0d5330")
        # ws['a1']='uigufyc'
        zhuan01=[]
        for i01 in data_list:
            n01 = re.sub(re.compile(r'[\000-\010]|[\013-\014]|[\016-\037]'), '', str(i01))

            if type(i01) == int or type(i01) == float:
                try:
                    n01 = int(n01)
                except:
                    try:
                        n01 = float(n01)
                    except:
                        pass

            zhuan01.append(n01)

        ws.append(zhuan01)
        wb.save(self.name01)
        print('数据已更新。。。。。。。。。。。。')


    # 对表进行修改，追加
    def xiubiao03(self,data_list):
        wb = openpyxl.load_workbook(self.name01)
        # 切换到目标数据表
        ws = wb[self.biaodan]
        for data in data_list:
            zhuan01=[]
            for i01 in data:
                n01=re.sub(re.compile(r'[\000-\010]|[\013-\014]|[\016-\037]'),'',str(i01))

                if type(i01) == int or type(i01) == float:
                    try:
                        n01 = int(n01)
                    except:
                        try:
                            n01 = float(n01)
                        except:
                            pass

                zhuan01.append(n01)

            ws.append(zhuan01)

        wb.save(self.name01)
        print('数据集已更新。。。。。。。。。。。。')
```
